```python
plt.subplots(figsize=(7, 3))
sns.countplot(x="season",data=matches, palette="magma")
plt.xticks(rotation=90)
plt.title('Total matches played in each season', fontsize = 10, fontweight = "bold")
plt.show()
```

```python
plt.figure(figsize=(20, 8))

ax = sns.countplot(x="venue", data=matches, palette="pastel")

ax.set_xlabel('Venue', fontsize=20, color='brown')
ax.set_ylabel('Count', fontsize=20)
ax.set_title('Total matches played in each Venue', fontsize=20, fontweight="bold")

plt.xticks(rotation=90, fontsize=20)
plt.yticks(fontsize=10)

plt.show()
```

```python
# df=pd.merge(matches,deliveries,on="id",how="left")
df = pd.merge(matches, deliveries, left_on="id", right_on="match_id", how="left")
df.info()
# df.info()
```

```python
season=df.groupby(['season'])['total_runs'].sum()
season_df = season.reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x="season", y="total_runs", data=season_df, palette="cividis")

plt.xlabel("Season", fontsize=14, color='brown')
plt.ylabel("Total Runs", fontsize=14)
plt.title("Total Runs per Season", fontsize=16, fontweight="bold")
plt.xticks(rotation=45)

plt.show()

```
